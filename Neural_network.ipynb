# Modelo de deep learning com keras
    A biblioteca keras utiliza como padrão a biblioteca tensorflow, porém, abstrai esta última, 
    tornando o trabalho muito menos complexo.
    A keras pode ser utilizada com diversos outros frameworks de deep learning.

```python
from sklearn import datasets
from sklearn .model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
```

```python
base = datasets.load_iris()
previsores = base.data
classe = base.target
classe
```

    Transformação da variável classe em dummy, ou seja, a realização da transformação da variável categórica classe, 
    em uma matriz na qual será indicado somente se a classe pertence ou não a cada categoria, sendo 1 = verdadeiro 
    e 0 = falso.

```python
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

    Divisão da variável classe_dummy em variáveis de treino e teste com 70% para o primeiro e 30% para o segundo.

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, classe_dummy, test_size= 0.3, random_state=0)
```

    Criação da estrutura do modelo da rede neural, com a sequencia de camadas.

```python
# Criação do objeto modelo, o qual é do tipo Sequential, o que será utilizado para para incluir as camadas uma após a outra.
modelo = Sequential()

#Primeira camada com 5 neurônios sendo 4 de entrada (um para cada característica das classes).
modelo.add(Dense(units=5, input_dim= 4))

# Segunda camada
modelo.add(Dense(units= 4))

#terceira camada com 3 neurônios, já que existem somente 3 classes e a função softmax, para lidar com mais de duas classes.
modelo.add(Dense(units= 3, activation= 'softmax'))
```

    Sumário com o resumo de como a rede foi construída

```python
modelo.summary()
```

    Utilização do método compile, utilizando os parâmetros mais comuns para otimizar os pesos da rede, sendo que
    o optimizer implementa o algorítmo adam, a loss function calcula o erro entre os valores previstos e os valores 
    reais e a métrica que irá medir o erro, também conhecida como taxa de acerto.

    Também a realização do treinamento do modelo, declarando quais variáveis deverão ser utilizadas para o treinamento 
    e quais para o teste, sendo que os dois são realizados em conjunto.

```python
modelo.compile(optimizer= 'adam', loss= 'categorical_crossentropy', metrics= ['accuracy'])
modelo.fit(x_treino, y_treino, epochs= 1000, validation_data= (x_teste, y_teste))
```

    Realização de uma previsão com o x_teste, levando em conta que a previsão retorna uma matriz de 0 e 1 foi realizada 
    a transformação em valores booleanos na segunda linha.

```python
previsoes= modelo.predict(x_teste)
previsoes= (previsoes > 0.5)
previsoes
```

    Para criar a matriz de confusão serão criadas duas variáveis, através de 
    loops, tendo em vista que o algoritmo possui três saídas será utilizada a função argmax do numpy, para pegar
    apenas o maior valor, qual seja, o valor True.

```python
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

    A matriz de confusão reflete exatamente o que a acurácia de 98% apresentada no modelo aponta.
    O teste apresentou apenas um erro, o que é muito superior aos resultados possíveis de serem obtidos
    com outros algorítimos que não envolvam deep learning.

```python
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```
